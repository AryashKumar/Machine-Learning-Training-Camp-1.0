```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
X=np.linspace(0,5,11)
```

```python
X
```

```python
y=X**2
```

```python
y
```

```python
plt.plot(X,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
plt.plot(y,X)
```

```python
plt.plot(X,y)
plt.plot(y,X)
```

```python
plt.subplot(1,3,1)  # nrows,ncolumns,index
plt.plot(X,y)
plt.subplot(1,3,2)
plt.plot(y,X)
plt.subplot(1,3,3)
plt.plot(y,X)
```

```python
fig1=plt.figure()
axes1=fig1.add_axes([0.1,0.3,0.8,1])  #(length,breath,width ,height)
axes2=fig1.add_axes([0.3,0.8,0.4,0.3])
axes1.plot(X,y)
axes2.plot(y,X)
```

```python
y_pi=np.array([1,2,4])
my_labels=['eren','yiu','demon']
plt.pie(y_pi,labels=my_labels)
```
